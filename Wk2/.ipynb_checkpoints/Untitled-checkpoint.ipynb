```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot(time,sequence,Type = '-',figsize= (10,6),label=None):
    plt.figure(figsize=figsize)
    plt.plot(time,sequence,Type,label=label)
    if label:
        plt.legend()
    plt.grid(True)
```

```python
def trend(time,slope=0):
    return time * slope
```

```python
time = np.arange(1,4 * 365 + 1)
plot(time,trend(time,slope=0.1))
```

```python
def seasonality(time,period,amplitude=1,phase=0):
    season = ((time + phase) % period) / period
    season = np.where(season < 0.4 , np.cos(2 * np.pi *season) , 1 / np.exp(3 * season))
    return season * amplitude
```

```python
time      = np.arange(1,4*365+1)
amplitude = 40
period    = 365
series    = seasonality(time,period=period,amplitude=amplitude,phase=0)

plot(time,series)
```

```python
time      = np.arange(1,4*365+1)
amplitude = 40
period    = 365
series    = seasonality(time,period=period,amplitude=amplitude,phase=0) + trend(time,slope=0.05)

plot(time,series)
```

```python
def noise(time,seed=42,scale=1):
    np.random.seed(seed)
    return np.random.randn(len(time)) * scale
```

```python
time      = np.arange(1,4*365+1)
series    = noise(time,seed=100,scale=.4)
plot(time,series)
```

```python
time      = np.arange(1,4*365+1)
amplitude = 40
period    = 365
series    = seasonality(time,period=period,amplitude=amplitude,phase=0) + trend(time,slope=0.05)\
            + noise(time,seed=100,scale=0.25)

plot(time,series)
```

```python
data = tf.data.Dataset.range(10)
data = data.window(size=5,shift=1,drop_remainder=True)
data = data.flat_map(lambda a : a.batch(5))
data = data.map(lambda a : (a[:-1],a[-1:]))
data = data.shuffle(10)
data = data.batch(2).prefetch(1)
for (x,y) in data:
    print(x.numpy())
    print(y.numpy())    
```

```python
def window_dataset(seriesies,window_size,batch_size,shuffle_buffer):
    dataset = tf.data.Dataset.from_tensor_slices(series)
    dataset = dataset.window(window_size+1,shift = 1,drop_remainder = True)
    dataset = dataset.flat_map(lambda a : a.batch(window_size+1))
    dataset = dataset.shuffle(shuffle_buffer)
    dataset = dataset.map(lambda a : (a[:-1],a[-1:]))
    dataset = dataset.batch(batch_size).prefetch(1)
    return dataset
```

```python
split_time = 1000
time_train = time[:split_time]
X_train    = series[:split_time]
time_valid = time[split_time:]
X_valid    = series[split_time:]
```

```python
window_size        = 20
batch_size         =  32
suffle_buffer_size = 1000
```

```python
dataset = window_dataset(series,window_size,batch_size,suffle_buffer_size)
```

```python
for x,y in dataset.take(1):
    print(y.numpy().shape)
```

```python
l0 = tf.keras.layers.Dense(1,input_shape=[window_size])
model = tf.keras.models.Sequential([l0])

model.compile(loss='mse',optimizer=tf.keras.optimizers.SGD(lr=1e-6,momentum=0.9))
history = model.fit(dataset,epochs=100,verbose=1)
```

```python
plt.figure(figsize=(10,6))
plt.plot(history.history['loss'],label="Loss")
plt.legend()
plt.grid(True)
```
